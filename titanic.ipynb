```python
df_gender = pd.read_csv("/kaggle/input/titanic/gender_submission.csv")
df_test = pd.read_csv("/kaggle/input/titanic/test.csv")
df_train = pd.read_csv("/kaggle/input/titanic/train.csv")
```

```python
df_train['Family'] = df_train['SibSp'] + df_train['Parch'] + 1
```

```python
cols = ["Pclass", "Sex", "Age","SibSp","Parch","Fare","Embarked","Family"]
X_train = df_train[cols]
y_train = df_train["Survived"]
```

```python
X_train.loc[:, 'Sex'] = X_train['Sex'].map({'male': 1, 'female': 2})

mean_age = X_train['Age'].mean()
X_train.loc[:, 'Age'] = X_train['Age'].fillna(mean_age)

most_embarked = df_train['Embarked'].mode()[0]
X_train.loc[:, 'Embarked'] = X_train['Embarked'].fillna(most_embarked)

X_train.loc[:, 'Embarked'] = X_train['Embarked'].map({'S': 3, 'C': 2, 'Q' : 1})

X_train.loc[:, 'Pclass'] = X_train['Pclass'].map({1: 100, 2: 50, 3 : 10})
X_train
```

```python
X_train.isna().sum()
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```
